```julia
using Plots
using DelimitedFiles
using ColorSchemes
using LaTeXStrings
```

```julia
file = raw"12_1917_0.0_probability_data.txt"
M = readdlm(file)
p_0_0 = M[:,1]; # eigenvalue index
p_x_bar_0 = M[:,2]; # level stat
iter_0 = M[:,3]; # level stat std

delta_1 = 0.01
file = raw"12_1917_0.01_probability_data.txt"
M = readdlm(file)
p_0_1 = M[:,1]; # eigenvalue index
p_x_bar_1 = M[:,2]; # level stat
iter_1 = M[:,3]; # level stat std

delta_2 =0.02
file = raw"12_1917_0.02_probability_data.txt"
M = readdlm(file)
p_0_2 = M[:,1]; # eigenvalue index
p_x_bar_2 = M[:,2]; # level stat
iter_2 = M[:,3]; # level stat std

delta_3 =0.03
file = raw"12_1917_0.03_probability_data.txt"
M = readdlm(file)
p_0_3 = M[:,1]; # eigenvalue index
p_x_bar_3 = M[:,2]; # level stat
iter_3 = M[:,3]; # level stat std

delta_4 =0.04
file = raw"12_1917_0.04_probability_data.txt"
M = readdlm(file)
p_0_4 = M[:,1]; # eigenvalue index
p_x_bar_4 = M[:,2]; # level stat
iter_4 = M[:,3]; # level stat std
```

```julia
colors = cgrad(:Set1_7,5, categorical = true)
colors = cgrad(:Set1_5,5, categorical = true)
```

```julia
gr()
L = 12;

MSW = 1
Linewidth_Solid  = 1.2
Linewidth_Dash   = 1.0
Markersize = 2
MarkerStrokeWidth = 0;
plot_font = "Computer Modern"
default(fontfamily=plot_font)

MyTitle = "L = "*string(L)*"; "*L"P(|0>):"*"dashed; "*L"P(|\bar{x}>)"*":solid"


p = plot(iter_1,p_0_1,linewidth=Linewidth_Solid,linecolor=colors[1],label = L"\delta = "*string(delta_1),)
p = plot!(iter_1,p_x_bar_1,linewidth=Linewidth_Dash,linecolor=colors[1],linestyle =:dash, label = "")

p = plot!(iter_2,p_0_2,linewidth=Linewidth_Solid,linecolor=colors[2],label = L"\delta = "*string(delta_2),)
p = plot!(iter_2,p_x_bar_2,linewidth=Linewidth_Dash,linecolor=colors[2],linestyle =:dash,label = "")

p = plot!(iter_3,p_0_3,linewidth=Linewidth_Solid,linecolor=colors[3],label = L"\delta = "*string(delta_3),)
p = plot!(iter_3,p_x_bar_3,linewidth=Linewidth_Dash,linecolor=colors[3],linestyle =:dash,label = "")

p = plot!(iter_4,p_0_4,linewidth=Linewidth_Solid,linecolor=colors[4],label = L"\delta = "*string(delta_4),)
p = plot!(iter_4,p_x_bar_4,linewidth=Linewidth_Dash,linecolor=colors[4],linestyle =:dash,label = "")

# green color.
p = plot!(iter_0,p_0_0,linewidth=Linewidth_Solid,linecolor=colors[5],label = L"\delta = 0.0",)
p = plot!(iter_0,p_x_bar_0,linewidth=Linewidth_Dash,linestyle =:dash,linecolor=colors[5],label = "",
    xlims = (0,250),
    titlefontsize=10,
    guidefontsize=8,
    tickfontsize=9,
    legendfontsize=8,
    title = L"L = 12, "*L"P_{0}(t) :"*"Soild, "*L"P_{\bar{x}}(t) :"*"Dash",
    framestyle = :box,
    thickness_scaling = 2.5,
    right_margin = Plots.mm,
    left_margin = Plots.mm,
    bottom_margin = Plots.mm,
    grid = false,
    dpi=600,)


plot!(size=(1200,900))


xlabel!("Time")
ylabel!("Probability")
#savefig(string(L)*"_probability_"*string(1917)*".png")
```
